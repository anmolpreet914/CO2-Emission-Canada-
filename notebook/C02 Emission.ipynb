```python
# Extract CO2 emissions data
co2_emissions = df['CO2 Emissions(g/km)']

# Calculate the first quartile (Q1)
Q1 = co2_emissions.quantile(0.25)

# Calculate the third quartile (Q3)
Q3 = co2_emissions.quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Calculate lower bound and upper bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
luxury_brands = ['ASTON MARTIN', 'AUDI', 'BENTLEY', 'BMW','BUICK', 'CADILLAC', 'CHEVROLET', 'INFINITI', 'JAGUAR', 'LEXUS', 'LINCOLN', 'MASERATI', 'MERCEDES-BENZ','PORSCHE', 'ROLLS-ROYCE', 'GENESIS']
df['IsLuxury'] = df['Make'].apply(lambda x: 1 if x in luxury_brands else 0)
```

```python
# Function to categorize model into different trim levels
def categorize_trim(model_name):
    model_name = model_name.lower()  # Convert model name to lowercase for case-insensitive matching
    
    # Keywords to identify different trim levels
    base_keywords = ['base']
    mid_level_keywords = ['lx', 'ex', 'le', 's']
    premium_keywords = ['premium', 'luxury', 'ultimate', 'platinum']
    sport_keywords = ['sport', 'si', 'rs', 'srt', 'trd', 'type r']
    off_road_keywords = ['off-road', 'trailhawk', 'fx4']
    special_edition_keywords = ['edition', 'special', 'limited', 'edition', 'anniversary']
    hybrid_keywords = ['hybrid']

    # Check for keywords in the model name
    if any(keyword in model_name for keyword in base_keywords):
        return 'Base Trim'
    elif any(keyword in model_name for keyword in mid_level_keywords):
        return 'Mid Level Trim'
    elif any(keyword in model_name for keyword in premium_keywords):
        return 'Premium Trim'
    elif any(keyword in model_name for keyword in sport_keywords):
        return 'Sport Trim'
    elif any(keyword in model_name for keyword in off_road_keywords):
        return 'Off Road Trim'
    elif any(keyword in model_name for keyword in special_edition_keywords):
        return 'Special Edition Trim'
    elif any(keyword in model_name for keyword in hybrid_keywords):
        return 'Hybrid Trim'
    else:
        return 'Other'  # If no specific trim level is identified

# Apply categorization function to create a new column for trim levels
df['Trim Level'] = df['Model'].apply(categorize_trim)

print(df)
```

```python
df['Vehicle Class'].unique()
```

```python
# Define mappings for grouping similar classes
class_mapping = {
    'SUV': ['SUV - SMALL', 'SUV - STANDARD', 'SUV - MID-SIZE'],
    'Van': ['MINIVAN', 'VAN - PASSENGER', 'VAN - CARGO'],
    'Truck': ['PICKUP TRUCK - STANDARD', 'PICKUP TRUCK - SMALL'],
    'Car': ['TWO-SEATER', 'STATION WAGON - SMALL', 'STATION WAGON - MID-SIZE', 'COMPACT', 'MID-SIZE', 'FULL-SIZE', 'MINICOMPACT', 'SUBCOMPACT'],
    'Special Purpose': ['SPECIAL PURPOSE VEHICLE']
}

# Function to map vehicle class to grouped class
def group_classes(vehicle_class):
    for group, classes in class_mapping.items():
        if vehicle_class in classes:
            return group
    return vehicle_class  # If the class doesn't match any group, return the original class

# Apply grouping to create a new column
df['Grouped Class'] = df['Vehicle Class'].apply(group_classes)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to 'Fuel Type'
df['Fuel Type Encoded'] = label_encoder.fit_transform(df['Fuel Type'])

# Apply label encoding to 'Transmission Type'
df['Transmission Type Encoded'] = label_encoder.fit_transform(df['Transmission Type'])

# Apply label encoding to grouped 'Vehicle Class'
df['Grouped Class Encoded'] = label_encoder.fit_transform(df['Grouped Class'])
```

```python
# Define the mapping of trim levels to numerical values
trim_level_mapping = {
    'Base Trim': 1,
    'Mid Level Trim': 2,
    'Premium Trim': 3,
    'Sport Trim': 4,
    'Off Road Trim': 5,
    'Special Edition Trim': 6,
    'Hybrid Trim': 7,
    'Other':8
}

# Apply ordinal encoding to 'Trim Level' using the mapping
df['Trim Level Encoded'] = df['Trim Level'].map(trim_level_mapping)
```

# Deploy the Model

```python
import numpy as np
import pandas as pd

# Generate random values for the independent variables
num_samples = 5  # Number of new data points to simulate
random_engine_size = np.random.uniform(1.0, 5.0, num_samples)  # Random float between 1.0 and 5.0
random_fuel_consumption = np.random.uniform(5.0, 15.0, num_samples)  # Random float between 5.0 and 15.0
random_gears = np.random.randint(4, 9, num_samples)  # Random integer between 4 and 8
random_is_luxury = np.random.randint(0, 2, num_samples)  # Random 0 or 1
random_fuel_type_encoded = np.random.randint(0, 5, num_samples)  # Random integers between 0 and 4
random_transmission_type_encoded = np.random.randint(0, 4, num_samples)  # Random integers between 0 and 3
random_grouped_class_encoded = np.random.randint(0, 4, num_samples)  # Random integers between 0 and 3
random_trim_level_encoded = np.random.randint(0, 7, num_samples)  # Random integers between 0 and 6

# Create a DataFrame with the random independent variables
random_ivs_df = pd.DataFrame({
    'Engine Size(L)': random_engine_size,
    'Fuel Consumption Comb (L/100 km)': random_fuel_consumption,
    'Gears': random_gears,
    'IsLuxury': random_is_luxury,
    'Fuel Type Encoded': random_fuel_type_encoded,
    'Transmission Type Encoded': random_transmission_type_encoded,
    'Grouped Class Encoded': random_grouped_class_encoded,
    'Trim Level Encoded': random_trim_level_encoded
})

print(random_ivs_df)
```

```python
# Load the trained model from the pickle file
model_filename = 'linear_regression_model.pkl'
with open(model_filename, 'rb') as file:
    loaded_model = pickle.load(file)

# Use the loaded model to predict the target variable (CO2 Emissions g/km) based on the random IVs
predictions = loaded_model.predict(random_ivs_df)

# Create a DataFrame with the random IVs and the predictions
result_df = random_ivs_df.copy()
result_df['Predicted CO2 Emissions(g/km)'] = predictions

# Display the result DataFrame
result_df.head()
```
